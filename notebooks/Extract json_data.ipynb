```python
import json 
import pandas as pd 
```

```python
def load_json(filepath):
    """
    --function to read json file 
    """
    try:
        with open(filepath) as file:
            json_file = json.load(file)
        return json_file
    except FileNotFoundError:
        print('this file is not found')
    
```

```python
data = load_json('../Data/global_design_data.json')
```

```python
data_list = list()

for game_key in data:
    for auto_id in data[game_key]:
        d = list()
        
        try: 
            d.append(game_key + '/' + auto_id)
            d.append(",".join(data[game_key][auto_id]['labels']['engagement']))
            d.append(",".join(data[game_key][auto_id]['text']['engagement']))
            d.append(",".join(data[game_key][auto_id]['colors']['engagement']))
            d.append(data[game_key][auto_id]['videos_data']['has_video'])
            d.append(data[game_key][auto_id]['eng_type']['eng_type'])
            d.append(data[game_key][auto_id]['direction']['direction'])
            d.append(data[game_key][auto_id]['adunit_sizes']['size_x'])
            d.append(data[game_key][auto_id]['adunit_sizes']['size_y'])
        except KeyError:
            d.append(None)
        
        data_list.append(d)

cols = ['game_key', 'labels', 'text', 'colors', 'video_data', 'eng_type','direction', 'adunit_size_x','adunit_size_y']
global_design_data = pd.DataFrame(data_list,columns=cols)
```
